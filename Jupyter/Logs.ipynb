```python
import csv
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

logs = pd.read_csv('tangibles_log.csv', index_col=False, sep=', ',  engine='python')
errorDrops = logs.loc[(logs["PID"] != "S1") & (logs["PID"] != "S2") & (logs["PID"] != "S3") & (logs["PID"] != "S4") & (logs["PID"] != "S5") & (logs["PID"] != "S6") & (logs["PID"] != "S7")].index
logsStamp = logs.drop(errorDrops) # droppt alle einträge bei denen die CSV datei nicht stimmt, dies beinhaltet die Maus Interaktionen aufgrund eines fehlenden "," im logging
logsStamp["CenterX"] = logsStamp["CenterX"].str[1:] #Erster character ist immer die klammer und die wird hier entfernt
logsStamp["CenterY"] = logsStamp["CenterY"].str[:-1] #Letzter Character wird entfernt, wieder die klammer


blues = ["midnightblue", "mediumblue", "royalblue", "cornflowerblue", "dodgerblue", "deepskyblue", "lightskyblue"]

# selektiert alle einträge die die mouse interaktion betreffen um sie zu berichtigen
mouseDrops = logs.loc[(logs["PID"] == "S1") | (logs["PID"] == "S2") | (logs["PID"] == "S3") | (logs["PID"] == "S4") | (logs["PID"] == "S5") | (logs["PID"] == "S6") | (logs["PID"] == "S7")].index
#Filtern aller Einträge der Mausevents aufgrund eines fehlenden "," im Log; so kann alles verwendet werden.
logsMouse = logs.drop(mouseDrops)
logsMouse["CenterY"] = logsMouse["CenterX"]
logsMouse["CenterX"] = logsMouse["Alive?"]
logsMouse["Alive?"] = logsMouse["ID"]
logsMouse["ID"] = logsMouse["Event-type"]
logsMouse["Event-type"] = logsMouse["Mode"]
logsMouse["Mode"] = logsMouse["Interaction Style"]
logsMouse["Interaction Style"] = logsMouse["PID"]
logsMouse["PID"] = logsMouse["Time"].str[-2:] # Korrektur der Fehler im logging
logsMouse["Time"] = logsMouse["Time"].str[:-3] # entfernen der letzten 3 character, also der PID und dem vorausgehenden leerzeichen
logsMouse["CenterX"] = logsMouse["CenterX"].str[1:] #Erster character ist immer die klammer und die wird hier entfernt
logsMouse["CenterY"] = logsMouse["CenterY"].str[:-1] #Letzter Character wird entfernt, wieder die klammer
logs = pd.concat([logsStamp, logsMouse])
logs["Alive?"] = logs["Alive?"].astype(bool)
logs.groupby("Interaction Style").describe()
```

```python
singleMouseLogs = logs.drop(logs.loc[logs["Interaction Style"] != "SM"].index)
singleMouseLogs.groupby("Mode").describe()
```

```python
logsScatter = logs.drop(logs.loc[logs["CenterX"] == "-500"].index)
logsScatter["CenterX"] = pd.to_numeric(logsScatter["CenterX"])
logsScatter["CenterY"] = pd.to_numeric(logsScatter["CenterY"])
```

```python
scatterSM = logsScatter.drop(logsScatter.loc[logsScatter["Interaction Style"] != "SM"].index)
scatterMM = logsScatter.drop(logsScatter.loc[logsScatter["Interaction Style"] != "MM"].index)
scatterSSt = logsScatter.drop(logsScatter.loc[logsScatter["Interaction Style"] != "SSt"].index)
scatterMSt = logsScatter.drop(logsScatter.loc[logsScatter["Interaction Style"] != "MSt"].index)


#Y-Achse wird in den jeweiligen plots invertiert da der nullpunkt des trackings die obere Linke ecke ist
plotSM = scatterSM.plot.scatter(x='CenterX', y='CenterY', c='DarkBlue', figsize = (15, 10), title="Single Mouse")
plotSM.invert_yaxis()
plotMM = scatterMM.plot.scatter(x='CenterX', y='CenterY', c='Blue',figsize = (15, 10), title="Multiple Mouse")
plotMM.invert_yaxis()
plotSSt = scatterSSt.plot.scatter(x='CenterX', y='CenterY', c='LightBlue',figsize = (15, 10), title="Single Stamp")
plotSSt.invert_yaxis()
plotMSt = scatterMSt.plot.scatter(x='CenterX', y='CenterY', c='Black',figsize = (15, 10), title="Multiple Stamp")
plotMSt.invert_yaxis()
```

# Log Ergebnisse
Die Logs zeichneten alle Bewegungen der Tangibles auf dem Tisch auf, sowie alles Schalten bei den Single Tangibles. 
Die Y-Achse der Scatterplots wurde invertiert um das Layout des Bildschirms korrekt darzustellen wie der Proband dazu saß. Aus den Graphen geht hervor, dass die Teilnehmer bei den Mouse Interaktionen beiweitem mehr Kontaktpunkte und Kontaktzeit zwischen Tisch und Tangible hatten, und viel der Interaktion auf dem Tisch selbst stattfand, während die Stamp Interaktionen generell weniger Kontaktpunkte hatten, zum Teil aufgrund der geringeren Gesamtzeit, aber zum Teil auch weil die Probanden bei der Stamp Interaktion generell weniger der Bewegung auf dem Tisch durchführten sondern meist nur aufsetzten um ein Bild einzusortieren und dann das Tangible wieder anzuheben. Die Scatterplots der Stamps geben noch sehr klare Rückschlüsse auf die Position der Bilder, während die der Mausinteraktion erlauben zu sehen in welchem Bereich des Bildschirms die meiste Interaktion stattfand, speziell bei Single Mouse ist gut sichtbar, dass die meiste Interaktion in der unteren Hälfte des Bildschirms stattfand, nahe am Teilnehmer.
